```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../csv2_/big_data_minmax_nm.csv')
```

```python
df.head()
```

```python
df = df.drop(columns='Unnamed: 0')
```

```python
def func_tier(x):
    if x>=5: return 1
    elif x>=2 and x<5: return 2
    elif x>=1 and x<2: return 3
    elif x>=0.5 and x<1: return 4
    else:   return 5
```

```python
df['tier'] = df['hit rate'].apply(lambda x:func_tier(x))
```

```python
df.head()
```

```python
df.to_csv('../csv2_/bit_data_minmax_tier.csv', index=False)
```

```python
df = pd.read_csv('../csv2_/big_data_robust_nm.csv')
df = df.drop(columns='Unnamed: 0')
df['tier'] = df['hit rate'].apply(lambda x:func_tier(x))
df.to_csv('../csv2_/bit_data_robust_tier.csv', index=False)
```
