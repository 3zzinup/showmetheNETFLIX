```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
import seaborn as sns
import openpyxl
import joblib
from sklearn.ensemble import RandomForestRegressor
```

```python
data = pd.read_csv('../csv2_/big_data_zscore.csv', header = 0)
data.head()

newData = data[['genre', 'rating', 'country', 'score', 'company', 'writer','director','hit rate']]
#속성(변수)선택
x = newData[['genre', 'rating', 'country', 'score', 'company', 'writer','director']]
y = newData[['hit rate']]

for i in (10, 20, 30, 40, 100, 150):
  model = RandomForestRegressor(n_estimators = i, random_state = 15)
  model.fit(x, y)

  relation_square = model.score(x, y)
  print('tree 개수 : ',i,'결정계수 : ', relation_square)
  
  y_p = model.predict(x)
  
  ax1 = sns.distplot(y, hist = False, label = 'y_real')
  ax1 = sns.distplot(y_p, hist = False, label = 'y_predict')

  ax1.legend(loc="best")
  ax1.set_xlim([-150, 150])
  plt.title("RandomForestRegression - predict hit rate")
  plt.xlabel("hit rate")
  plt.ylabel("distribution")
  plt.show()
  if i == 10:
    plt.savefig("../_graph/RFR_tree100_h.jpeg", dpi=300)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from mlxtend.plotting import plot_decision_regions
import numpy as np
import pandas as pd
import time
import csv
```

```python
df = pd.read_csv('../csv2_/big_data_zscore_tier.csv')

newData = df[['genre', 'rating', 'country', 'score', 'company', 'writer','director','tier']]
#속성(변수)선택
X = newData[['genre', 'rating', 'country', 'score', 'company', 'writer','director']]
y = newData[['tier']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

ml = RandomForestClassifier(criterion='entropy', n_estimators=10)

ml.fit(X_train, y_train)

y_pred = ml.predict(X_test)

y_pred = y_pred.reshape(1627,1)
joblib.dump(ml, '../models_/rfc_model_z_t.pkl')
print('총 테스트 개수 : %d, 오류개수 : %d'%(len(y_test), (y_test!=y_pred).sum()))
print('정확도 : %.2f'%accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
from sklearn.metrics import accuracy_score 
print ("Accuracy : ", accuracy_score(y_test, y_pred))
cm
```

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt

def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.2):
    # 마커와 컬러맵 설정하기
    markers=('s', 'x', 'o', '^', 'v')
    colors=('red','blue','lightgreen','gray','cyan')
    cmap=ListedColormap(colors[:len(np.unique(y))])
    
    # 결정경계를 그리기
    x1_min, x1_max = X[:,0].min()-1, X[:,0].max()+1
    x2_min, x2_max = X[:,1].min()-1, X[:,1].max()+1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                          np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y==cl,0], y=X[y==cl,1],
                   alpha=0.8, c=colors[idx],
                   marker=markers[idx], label=cl,
                   edgecolor='black')
        
    # 테스트 샘플을 부각하여 그리기
    if test_idx:
        X_test, y_test = X[test_idx, :], y[test_idx]
        plt.scatter(X_test[:,0], X_test[:,1],
                   c='', edgecolor='black',alpha=1.0,
                   linewidth=1, marker='o',
                   s=100, label='test set')
```

```python
from sklearn.svm import SVR
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../csv2_/big_data_zscore_tier.csv')

newData = df[['genre', 'rating', 'country', 'score', 'company', 'writer','director','tier']]
#속성(변수)선택
X = newData[['genre', 'rating', 'country', 'score', 'company', 'writer','director']]
y = newData[['tier']]


for i in (0.1, 1, 10, 100, 1000):
  for j in (0.1, 1, 10, 100, 1000):
    model = SVR(kernel = 'rbf', C=i, gamma = j)
    model.fit(X, y)
    relation_square = model.score(X, y)
    print('결정계수 R : ', relation_square)

y_p = model.predict(X)
joblib.dump(model, '../models_/svm_model_z_t.pkl')
ax1 = sns.distplot(y, hist = False, label = 'y_real')
ax2 = sns.distplot(y_p, hist = False, label = 'y_predict')
ax1.legend(loc="best")

plt.title("SVM(Support Vector Model) - predict tier")
plt.xlabel("tier")
plt.ylabel("distribution")
plt.show()
```
